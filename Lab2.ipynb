# Lab 2: Creación de Usuarios y Políticas IAM (Admin y Practicante)

**Objetivo:**

En este laboratorio crearás dos usuarios IAM.

- El primer usuario será un "usuario administrador" que utilizarás para el resto de los laboratorios, con permisos de administrador utilizando una política administrada por AWS.
- El segundo usuario será un "usuario practicante" al que se le denegarán todas las acciones de EC2 mediante una política personalizada creada con el editor visual.

Al final, se validará el acceso del usuario practicante desde una ventana de incógnito en el navegador para comprobar la restricción sobre EC2.

---

Pasos a Seguir:

1. Acceder a la Consola de AWS:
    - Abre tu navegador y ve a: [https://console.aws.amazon.com](https://console.aws.amazon.com/)
    - Inicia sesión con tus credenciales de la cuenta principal (Root).
2. Ir al Servicio IAM:
    - En el buscador de servicios, escribe "IAM" y selecciona "IAM".

---

**Crear el Usuario Administrador**

1. Crear Usuario Administrador:
    - En el menú izquierdo de IAM, selecciona "Users".
    - Haz clic en "Add users".
    - User name: "usuario-admin".
    - Access type:
        - Selecciona "AWS Management Console access" y luego / I want to create an IAM user
        - Selecciona Custom Password y crea una contraseña para el usuario.
    - Haz clic en "Next".
2. Asignar Permisos al Usuario Administrador:
    - En "Permissions", selecciona "Attach policies directly".
    - Busca en el search bar debajo de Permission Policies la política administrada por AWS llamada "AdministratorAccess" y márcala.
    - Haz clic en "Next"
    - Revisa los detalles y haz clic en "Create user".

---

**Crear el Usuario Practicante con Política Personalizada**

1. Crear Usuario Practicante:
    - En el menú izquierdo, selecciona "Users".
    - Haz clic en "Add users".
    - User name: "usuario-practicante".
    - Access type:
        - Marca la opción
        - [ ]  Provide user access to the AWS Management Console
        
        y luego seleccion I want to create an IAM user
        
        - Establece una contraseña y opcionalmente forzar el cambio en el primer inicio.
    - Haz clic en "Next".
    
    (No adjuntes políticas aún, primero crearemos la política personalizada).
    
2. **Crear la Política Personalizada con el Editor Visual para Denegar EC2:**
    1. Selecciona la opcion Attach policies directly
    - En la sección de permisos, haz clic en "Create policy" (se abrira una nueva pestaña).
    - En la pestaña "Visual editor":
        - Selecciona el servicio: "EC2".
        - Cambia "Effect": "Allow" por "Effect": "Deny".
        - En Actions, selecciona "All EC2 actions" (todas las acciones).
        - En Resources, deja "All resources".
    
    Revisa la política y haz clic en "Next".
    
    - Name: "DenyAllEC2Actions"
    - Description: "Deniega todas las acciones sobre EC2"
    - Haz clic en "Create policy".
    
    Cierra la pestaña de la política y regresa a la pestaña de creación del usuario.
    
3. Adjuntar la Política Personalizada al Usuario Practicante:
    - Haz clic en "Refresh" en la sección de permisos.
    - En "Attach existing policies directly", busca "DenyAllEC2Actions".
    - Selecciónala.
    - Haz clic en "Next".
    - Revisa y haz clic en "Create user".

---

**Verificación del Acceso del Usuario Practicante**

1. Abre una Ventana de Incógnito en tu navegador:
    - Este paso es importante para evitar conflictos con tu sesión actual.
2. Inicia Sesión con el Usuario Practicante:
    - Ve a la URL de la consola de AWS: [https://console.aws.amazon.com](https://console.aws.amazon.com/)
    - Ingresa el nombre de cuenta o alias, el "User name" (usuario-practicante) y la contraseña.
3. Probar Acceso a EC2:
    - Desde la consola del usuario practicante, ve a "EC2".
    - Intenta listar instancias o crear una instancia.
    - Deberías ver errores o denegación de acceso indicando que la política prohíbe las acciones de EC2.
4. Validar que el Usuario Admin sí Tiene Acceso a EC2:
    - En otra ventana (donde tienes la sesión con tu usuario principal o inicia sesión con el usuario-admin) verifica que puedes acceder y administrar EC2 sin problemas.

---

**Consejos de Seguridad**

- Principio de privilegios mínimos: el usuario practicante no puede acceder a EC2.
- El usuario administrador cuenta con todos los privilegios gracias a "AdministratorAccess".
- Puedes rotar credenciales periódicamente y usar MFA.

---

**Fin del Lab.**

Este laboratorio demuestra cómo crear usuarios con diferentes niveles de acceso, asignar políticas administradas por AWS y crear una política personalizada con el editor visual para denegar permisos sobre un servicio específico, validando en una ventana de incógnito el resultado
