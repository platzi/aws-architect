# **Lab 8: Configuración de Versionamiento en S3**

En este laboratorio, aprenderás a habilitar el **versionamiento** en un bucket de S3, subir objetos, modificar archivos y verificar cómo el versionamiento permite mantener un historial de cambios para los objetos almacenados.

---

**Objetivos del laboratorio**

1. Crear un bucket de S3 y habilitar el versionamiento.
2. Subir y modificar archivos para generar múltiples versiones.
3. Visualizar y gestionar versiones de objetos en el bucket.
4. Recuperar una versión específica de un objeto.

---

**Instrucciones detalladas**

**1. Crear un bucket de S3**

1. Ve al servicio **Amazon S3** en la consola de AWS.
2. Haz clic en **Create bucket**.
3. Configura el bucket:
    - **Bucket name**: `lab8-versioning-bucket-<tu-nombre>` (elige un nombre único).
    - **AWS Region**: Elige la región donde deseas crear el bucket.
    - En las opciones predeterminadas, asegúrate de **habilitar el bloqueo de acceso público** para fines de este laboratorio.
4. Haz clic en **Create bucket**.

---

**2. Habilitar el versionamiento**

1. En la lista de buckets, selecciona el bucket recién creado (`lab7-versioning-bucket-<tu-nombre>`).
2. Ve a la pestaña **Properties**.
3. Busca la sección **Bucket Versioning** y haz clic en **Edit**.
4. Activa la opción **Enable** y haz clic en **Save changes**.

---

**3. Subir un archivo inicial**

1. Ve a la pestaña **Objects** dentro de tu bucket.
2. Haz clic en **Upload** y selecciona un archivo (por ejemplo, un archivo de texto llamado `example.txt`).
3. Haz clic en **Upload** para completar la carga.

---

**4. Modificar el archivo para crear versiones**

1. Crea una copia modificada del archivo original en tu computadora (por ejemplo, cambia el contenido de `example.txt`).
2. Regresa a la consola de S3 y vuelve a subir el archivo `example.txt` al bucket.
3. S3 mostrará un mensaje indicando que estás sobrescribiendo un archivo existente.
4. Haz clic en **Upload** para reemplazar el archivo.

---

**5. Visualizar las versiones del archivo**

1. En la pestaña **Objects**, selecciona **Show versions** (arriba a la derecha).
2. Ahora verás una lista de todas las versiones del archivo `example.txt`, identificadas por:
    - **Key**: Nombre del archivo.
    - **Version ID**: Identificador único para cada versión.
    - **Last modified**: Fecha y hora de la modificación.

---

**6. Recuperar una versión específica**

1. Identifica la versión anterior del archivo desde la lista de versiones.
2. Selecciona esa versión y haz clic en **Download** para descargarla a tu computadora.
3. Verifica que el contenido corresponde a la versión seleccionada.

---

**7. Eliminar un archivo con versionamiento habilitado**

1. Elimina el archivo `example.txt`:
    - Selecciónalo y haz clic en **Delete**.
    - Confirma la eliminación.
2. Regresa a la lista de versiones y observa que el archivo no se elimina por completo:
    - Se crea un marcador de eliminación (*delete marker*).
3. Para recuperar el archivo, selecciona el marcador de eliminación y elimínalo. El archivo original reaparecerá.

---

**Resultados esperados**

- El bucket tiene habilitado el versionamiento.
- Cada vez que se sube una nueva versión de un archivo, se almacena como una versión independiente.
- Las versiones anteriores pueden visualizarse, descargarse o restaurarse.
- Los objetos eliminados pueden recuperarse eliminando el marcador de eliminación.

---

**Conclusión**

El versionamiento en S3 es una característica poderosa para preservar el historial de los objetos y proteger contra eliminaciones accidentales. Este laboratorio muestra cómo usar esta funcionalidad para administrar cambios en los datos almacenados en S3 de manera efectiva.
